```python
# 0(n)
def square_sum(n):
    result = 0
    for i in range(1,n+1):
        result += i**2
    return result
```

```python
import time

start_time = time.time()

square_sum(10000000)

end_time = time.time()
elapsed_time_1 = end_time - start_time

print(f"Time taken 1.Function: {elapsed_time_1}")
```

```python
# O(1)
def square_sum2(n):
    return (n*(n+1)*(2*n+1)) / 6
```

```python
import time

start_time = time.time()

square_sum2(10000000)

end_time = time.time()
elapsed_time_2 = end_time - start_time

print(f"Time taken 2.Function: {elapsed_time_2}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

diff = elapsed_time_1 - elapsed_time_2

sns.barplot(x=['Function1', 'Function2'], y=[elapsed_time_1, elapsed_time_2])
plt.title('Difference: {}'.format(diff))

plt.show()
```

# Big-O Notation

```python
# Constant O(1)
# Input = n          --> Output = 1

def constant_bigO(liste):
    print(liste[1])
```

```python
constant_bigO([3,4,6,7,218,57,26]) # Constant
```

```python
# Linear O(n)
# Input = n          --> Output = n

def linear_bigO(liste):
  for i in liste:
    print(i)
```

```python
linear_bigO([3,4,6,7,218,57,26]) # Linear
```

```python
# Cubic O(n^3)
# Input = n          --> Output = n**3
```

```python
def cubic_bigO(liste): # O(n^3)
  for i in liste:
    for j in liste:
      for k in liste:
        print(i, j, k)
```

```python
cubic_bigO([3,4,6,7,218,57,26])
```

```python
# Calculating Scale Of Big-O
# Insignificant constant
# O(n) = linear
# O(2n) = linear          # 2 is insignificant constant and this is still linear.
```

```python
def linear_big_o2(liste): # O(2n)
  for i in liste:
    print(i)
  for i in liste:
    print(i)
```

```python
linear_big_o2([3,4,6,7,218,57,26]) 
```

```python
# O(1+n)

def example(liste):
  print(liste[0]) # O(1)

  for j in liste: # O(n)
    print(j) 
```

```python
example([3,4,6,7,218,57,26]) 
```
